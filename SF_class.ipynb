```python
import pandas as pd
```

```python
metadata  = pd.read_csv('mjd_of_maximum_brightness.csv')
band_peak = np.array(metadata['band_peak'])
```

```python
V,B,g,R,r,unknown,U,u =0,0,0,0,0,0,0,0

for i in band_peak:
    if str(i) == 'V':
        V+=1
    elif str(i) == 'B':
        B+=1
    elif str(i) == 'g':
        g+=1
    elif str(i) == 'R':
        R+=1
    elif str(i) == 'r':
        r+=1
    elif str(i) == band_peak[-2]:
        unknown+=1
    elif str(i) == 'U':
        U+=1
    elif str(i) == 'u':
        u+=1       
    else:
        print(i)
        
```
